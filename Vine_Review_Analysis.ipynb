```python
import pandas as pd
```

```python
# Read in file
file = "vine_table.csv"
raw_df = pd.read_csv(file)
raw_df.head()
```

```python
# Filter vine_df for 20 or more total_votes
filtered_df = raw_df.loc[(raw_df['total_votes'] >= 20) & 
    (raw_df['helpful_votes'] / raw_df['total_votes'] >= 0.5)]
filtered_df.tail()
```

```python
# Filter for vine reviews only
vine_df = filtered_df.loc[filtered_df['vine'] == 'Y']
vine_df
```

```python
# Filter for non-vine reviews only
nonvine_df = filtered_df.loc[filtered_df['vine'] == 'N']
nonvine_df
```

```python
# Our Analysis for Vine

v_total_reviews = len(vine_df['review_id'])
v_total_5star = vine_df['star_rating'].loc[vine_df['star_rating'] == 5].count()
v_pct_5star = v_total_5star / v_total_reviews

print(f' Out of {v_total_reviews} reviews of Vine products, {v_total_5star} or {v_pct_5star*100}% were 5-star reviews.')
```

```python
# Our Analysis for Non-Vine

nv_total_reviews = len(nonvine_df['review_id'])
nv_total_5star = nonvine_df['star_rating'].loc[nonvine_df['star_rating'] == 5].count()
nv_pct_5star = nv_total_5star / nv_total_reviews

print(f' Out of {nv_total_reviews} reviews of non-Vine products, {nv_total_5star} or {nv_pct_5star*100}% were 5-star reviews.')
```
